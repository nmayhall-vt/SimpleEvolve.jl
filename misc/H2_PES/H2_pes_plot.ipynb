```julia
using JLD2
energy_list=[]
fci_energy_list=[]
error_list=[]
for dist in 0.5:0.05:3.00
    filename = "final_results_H2_$(dist).jld2"
    data = load(filename)
    display(data)
    push!(energy_list,data["energies"])
    push!(fci_energy_list,data["fci_energies"])
    push!(error_list,data["errors"])
end
```

```julia
display(fci_energy_list)
display(energy_list)
display(error_list)
```

```julia
using Plots

n_samples = length(energy_list)
distances = [round(x, digits=2) for x in 0.5:0.05:3.0]

# Extract ground state energies as vectors
ground_state_energies = [energy_list[i][1] for i in 1:n_samples]
fci_ground_energies = [fci_energy_list[i][1] for i in 1:n_samples]

# Plot
plot(distances, ground_state_energies, lw=2, label="VQE Ground State")
plot!(distances, fci_ground_energies, lw=2, label="FCI Ground State")
xlabel!("Distance (Å)")
ylabel!("Energy (Ha)")
title!("Ground State Energy vs Distance")

```

```julia
using Plots

# Your data setup
distances = [round(x, digits=2) for x in 0.5:0.05:3.0]  # 51 elements
n_samples = length(distances)  # 51
n_states = 4  # Ground + 3 excited states

# Initialize plot
plot(title="Energy vs Distance", xlabel="Distance (Å)", ylabel="Energy (Ha)")

# Plot each state individually
for j in 1:n_states
    # Extract energies for state j across all 51 distances
    vqe_energies = [energy_list[i][j] for i in 1:n_samples]
    fci_energies = [fci_energy_list[i][j] for i in 1:n_samples]
    
    # Plot VQE and FCI for this state
    plot!(distances, vqe_energies, lw=2, label="VQE State $j")
    plot!(distances, fci_energies, lw=2, linestyle=:dash,marker="o", label="FCI State $j")
end

display(plot!())
```
